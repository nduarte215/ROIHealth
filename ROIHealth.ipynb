```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>CuraData - Your Bloodwork Reveals Hidden Productivity Costs</title>
    <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
    <style>
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }

        body {
            font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, sans-serif;
            background: #0f0f23;
            color: #e2e8f0;
            line-height: 1.6;
        }

        /* Hero Section */
        .hero {
            min-height: 100vh;
            display: flex;
            align-items: center;
            justify-content: center;
            position: relative;
            background: linear-gradient(135deg, #0f0f23 0%, #1a1a3e 100%);
            overflow: hidden;
        }

        .hero::before {
            content: '';
            position: absolute;
            top: 0;
            left: 0;
            right: 0;
            bottom: 0;
            background: radial-gradient(circle at 20% 50%, rgba(120, 80, 255, 0.3) 0%, transparent 50%);
            animation: float 20s ease-in-out infinite;
        }

        @keyframes float {
            0%, 100% { transform: translate(0, 0); }
            50% { transform: translate(30px, -30px); }
        }

        .hero-content {
            max-width: 1200px;
            padding: 2rem;
            z-index: 10;
            position: relative;
            text-align: center;
        }

        .hero h1 {
            font-size: clamp(2.5rem, 5vw, 4rem);
            font-weight: 800;
            margin-bottom: 1.5rem;
            background: linear-gradient(135deg, #60a5fa 0%, #c084fc 50%, #f472b6 100%);
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
            background-clip: text;
        }

        .impact-number {
            font-size: 4rem;
            font-weight: 900;
            color: #f59e0b;
            text-shadow: 0 0 30px rgba(245, 158, 11, 0.5);
        }

        .hero-subtitle {
            font-size: 1.5rem;
            color: #94a3b8;
            margin: 2rem 0;
        }

        .biomarker-examples {
            display: flex;
            justify-content: center;
            gap: 3rem;
            margin: 3rem 0;
            flex-wrap: wrap;
        }

        .biomarker-card {
            background: rgba(255, 255, 255, 0.05);
            backdrop-filter: blur(10px);
            padding: 1.5rem;
            border-radius: 1rem;
            border: 1px solid rgba(255, 255, 255, 0.1);
            min-width: 200px;
        }

        .biomarker-card h3 {
            color: #60a5fa;
            margin-bottom: 0.5rem;
        }

        .biomarker-card p {
            color: #cbd5e1;
            font-size: 0.9rem;
        }

        .cta-button {
            background: linear-gradient(135deg, #6366f1 0%, #8b5cf6 100%);
            color: white;
            border: none;
            padding: 1.5rem 3rem;
            font-size: 1.2rem;
            font-weight: 600;
            border-radius: 0.5rem;
            cursor: pointer;
            transition: all 0.3s ease;
            box-shadow: 0 10px 25px rgba(99, 102, 241, 0.3);
            margin-top: 2rem;
        }

        .cta-button:hover {
            transform: translateY(-2px);
            box-shadow: 0 15px 35px rgba(99, 102, 241, 0.4);
        }

        /* Analysis Section */
        .analysis-section {
            max-width: 1200px;
            margin: 0 auto;
            padding: 4rem 2rem;
        }

        .section-header {
            text-align: center;
            margin-bottom: 3rem;
        }

        .section-header h2 {
            font-size: 2.5rem;
            margin-bottom: 1rem;
            background: linear-gradient(135deg, #60a5fa 0%, #c084fc 100%);
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
            background-clip: text;
        }

        .glass-panel {
            background: rgba(255, 255, 255, 0.05);
            backdrop-filter: blur(10px);
            border: 1px solid rgba(255, 255, 255, 0.1);
            border-radius: 1rem;
            padding: 2rem;
            margin-bottom: 2rem;
        }

        .input-group {
            margin-bottom: 1.5rem;
        }

        .input-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: #cbd5e1;
            text-transform: uppercase;
            font-size: 0.875rem;
            letter-spacing: 0.05em;
        }

        .input-group textarea {
            width: 100%;
            min-height: 200px;
            padding: 1rem;
            background: rgba(255, 255, 255, 0.05);
            border: 1px solid rgba(255, 255, 255, 0.2);
            border-radius: 0.5rem;
            color: #e2e8f0;
            font-family: 'Monaco', 'Courier New', monospace;
            resize: vertical;
            transition: all 0.3s ease;
        }

        .input-group textarea:focus {
            outline: none;
            border-color: #60a5fa;
            background: rgba(255, 255, 255, 0.08);
        }

        .input-options {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
            gap: 1rem;
            margin-bottom: 2rem;
        }

        .input-group input,
        .input-group select {
            width: 100%;
            padding: 0.75rem;
            background: rgba(255, 255, 255, 0.05);
            border: 1px solid rgba(255, 255, 255, 0.2);
            border-radius: 0.5rem;
            color: #e2e8f0;
        }

        .sample-format {
            background: rgba(99, 102, 241, 0.1);
            border: 1px solid rgba(99, 102, 241, 0.3);
            padding: 1rem;
            border-radius: 0.5rem;
            margin-bottom: 1.5rem;
        }

        .sample-format h4 {
            color: #60a5fa;
            margin-bottom: 0.5rem;
        }

        .sample-format pre {
            color: #cbd5e1;
            font-size: 0.9rem;
        }

        /* Results Section */
        .results-section {
            display: none;
            margin-top: 3rem;
        }

        .impact-summary {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(250px, 1fr));
            gap: 1.5rem;
            margin-bottom: 2rem;
        }

        .impact-card {
            background: linear-gradient(135deg, rgba(99, 102, 241, 0.1) 0%, rgba(139, 92, 246, 0.1) 100%);
            padding: 1.5rem;
            border-radius: 1rem;
            border: 1px solid rgba(99, 102, 241, 0.2);
            text-align: center;
            transition: all 0.3s ease;
        }

        .impact-card:hover {
            transform: scale(1.05);
            box-shadow: 0 10px 30px rgba(99, 102, 241, 0.2);
        }

        .impact-icon {
            font-size: 2.5rem;
            margin-bottom: 1rem;
        }

        .impact-value {
            font-size: 2rem;
            font-weight: 800;
            color: #60a5fa;
            margin-bottom: 0.5rem;
        }

        .impact-label {
            color: #94a3b8;
            font-size: 0.875rem;
            text-transform: uppercase;
        }

        .chart-container {
            height: 400px;
            margin-bottom: 2rem;
        }

        .recommendations {
            background: linear-gradient(135deg, rgba(34, 197, 94, 0.1) 0%, rgba(16, 185, 129, 0.1) 100%);
            border: 1px solid rgba(34, 197, 94, 0.3);
            padding: 2rem;
            border-radius: 1rem;
        }

        .recommendation-item {
            padding: 1rem;
            margin-bottom: 1rem;
            background: rgba(255, 255, 255, 0.05);
            border-radius: 0.5rem;
            border-left: 4px solid #22c55e;
        }

        .loading {
            display: none;
            text-align: center;
            padding: 3rem;
        }

        .loading-spinner {
            width: 50px;
            height: 50px;
            border: 3px solid rgba(99, 102, 241, 0.3);
            border-top-color: #6366f1;
            border-radius: 50%;
            animation: spin 1s linear infinite;
            margin: 0 auto 1rem;
        }

        @keyframes spin {
            to { transform: rotate(360deg); }
        }

        /* How it works section */
        .how-it-works {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(300px, 1fr));
            gap: 2rem;
            margin: 3rem 0;
        }

        .step {
            text-align: center;
            padding: 2rem;
        }

        .step-number {
            display: inline-block;
            width: 50px;
            height: 50px;
            background: linear-gradient(135deg, #6366f1 0%, #8b5cf6 100%);
            border-radius: 50%;
            line-height: 50px;
            font-size: 1.5rem;
            font-weight: bold;
            margin-bottom: 1rem;
        }

        .step h3 {
            color: #e2e8f0;
            margin-bottom: 0.5rem;
        }

        .step p {
            color: #94a3b8;
        }
    </style>
</head>
<body>
    <!-- Hero Section -->
    <section class="hero">
        <div class="hero-content">
            <h1>Your Bloodwork Reveals Hidden Productivity Costs</h1>
            <p class="hero-subtitle">
                American businesses lose <span class="impact-number">$530 BILLION</span> annually
                from undiagnosed cognitive dysfunction
            </p>

            <div class="biomarker-examples">
                <div class="biomarker-card">
                    <h3>Low Vitamin D</h3>
                    <p>‚Üí $15,000/year lost productivity</p>
                    <p>‚Üí 65% have deficiency</p>
                </div>
                <div class="biomarker-card">
                    <h3>High Inflammation</h3>
                    <p>‚Üí 8+ sick days/year</p>
                    <p>‚Üí 2x turnover risk</p>
                </div>
                <div class="biomarker-card">
                    <h3>Immune Dysfunction</h3>
                    <p>‚Üí 40% cognitive decline</p>
                    <p>‚Üí Team contagion effect</p>
                </div>
            </div>

            <button class="cta-button" onclick="scrollToAnalysis()">
                Analyze Your Bloodwork Now
            </button>
        </div>
    </section>

    <!-- Analysis Section -->
    <section class="analysis-section" id="analysis">
        <div class="section-header">
            <h2>Turn Your Lab Results Into Business Intelligence</h2>
            <p style="color: #94a3b8; font-size: 1.2rem;">
                Upload your bloodwork to see the real cost of biomarker imbalances
            </p>
        </div>

        <!-- How it Works -->
        <div class="glass-panel">
            <div class="how-it-works">
                <div class="step">
                    <div class="step-number">1</div>
                    <h3>Upload Lab Results</h3>
                    <p>Paste your bloodwork data or upload a PDF</p>
                </div>
                <div class="step">
                    <div class="step-number">2</div>
                    <h3>AI Analysis</h3>
                    <p>We identify cognitive risk factors</p>
                </div>
                <div class="step">
                    <div class="step-number">3</div>
                    <h3>Get ROI Report</h3>
                    <p>See productivity costs & solutions</p>
                </div>
            </div>
        </div>

        <!-- Input Panel -->
        <div class="glass-panel">
            <h3 style="margin-bottom: 1.5rem; color: #e2e8f0;">Enter Your Bloodwork Data</h3>

            <div class="sample-format">
                <h4>üìã Supported Format:</h4>
                <pre>Vitamin D: 25 ng/mL
Neutrophils: 65%
Lymphocytes: 25%
WBC: 6.5 x10¬≥/¬µL</pre>
            </div>

            <div class="input-group">
                <label for="labReport">Paste Your Lab Results</label>
                <textarea id="labReport" placeholder="Copy and paste your lab results here...

Example:
Complete Blood Count (CBC)
Date: 01/15/2024

Vitamin D (25-OH): 22 ng/mL [Low]
WBC: 7.2 x10¬≥/¬µL
Neutrophils: 68%
Lymphocytes: 22%"></textarea>
            </div>

            <div class="input-options">
                <div class="input-group">
                    <label for="companySize">Company/Team Size</label>
                    <input type="number" id="companySize" value="1" min="1">
                </div>
                <div class="input-group">
                    <label for="role">Your Role</label>
                    <select id="role">
                        <option value="individual">Individual Contributor</option>
                        <option value="knowledge_worker">Knowledge Worker</option>
                        <option value="manager">Manager</option>
                        <option value="executive">Executive</option>
                    </select>
                </div>
            </div>

            <div style="display: flex; gap: 1rem; justify-content: center;">
                <button class="cta-button" onclick="analyzeBloodwork()">
                    Calculate Business Impact
                </button>
                <button class="cta-button" style="background: transparent; border: 2px solid #60a5fa; color: #60a5fa;" onclick="loadSampleData()">
                    Try Sample Data
                </button>
            </div>
        </div>

        <!-- Loading State -->
        <div class="loading" id="loading">
            <div class="loading-spinner"></div>
            <p style="color: #94a3b8;">Analyzing biomarkers and calculating productivity impact...</p>
        </div>

        <!-- Results -->
        <div class="results-section" id="results">
            <div class="glass-panel">
                <h3 style="margin-bottom: 1.5rem; color: #e2e8f0;">Your Productivity Impact Report</h3>

                <div class="impact-summary" id="impactSummary">
                    <!-- Impact cards will be inserted here -->
                </div>

                <div class="chart-container">
                    <canvas id="costChart"></canvas>
                </div>

                <div class="chart-container">
                    <canvas id="roiChart"></canvas>
                </div>
            </div>

            <div class="recommendations" id="recommendations">
                <h3 style="margin-bottom: 1.5rem; color: #e2e8f0;">Your Personalized Action Plan</h3>
                <!-- Recommendations will be inserted here -->
            </div>
        </div>
    </section>

    <script>
        let costChart = null;
        let roiChart = null;

        // Configure Chart.js defaults
        Chart.defaults.color = '#94a3b8';
        Chart.defaults.borderColor = 'rgba(148, 163, 184, 0.1)';

        function scrollToAnalysis() {
            document.getElementById('analysis').scrollIntoView({
                behavior: 'smooth'
            });
        }

        async function analyzeBloodwork() {
            const labText = document.getElementById('labReport').value;
            const companySize = parseInt(document.getElementById('companySize').value) || 1;
            const role = document.getElementById('role').value;

            if (!labText.trim()) {
                alert('Please enter your lab results');
                return;
            }

            // Show loading
            document.getElementById('loading').style.display = 'block';
            document.getElementById('results').style.display = 'none';

            try {
                const response = await fetch('/api/business-health-impact', {
                    method: 'POST',
                    headers: {
                        'Content-Type': 'application/json',
                    },
                    body: JSON.stringify({
                        lab_text: labText,
                        company_size: companySize,
                        role: role
                    })
                });

                const data = await response.json();

                if (data.success) {
                    displayResults(data);
                } else {
                    alert('Analysis failed: ' + (data.error_message || 'Unknown error'));
                }
            } catch (error) {
                alert('Error: ' + error.message);
            } finally {
                document.getElementById('loading').style.display = 'none';
            }
        }

        function displayResults(data) {
            const impact = data.business_impact;

            // Display impact cards
            const impactHTML = `
                <div class="impact-card">
                    <div class="impact-icon">üí∞</div>
                    <div class="impact-value">${impact.key_metrics.annual_cost}</div>
                    <div class="impact-label">Annual Productivity Cost</div>
                </div>
                <div class="impact-card">
                    <div class="impact-icon">‚è±Ô∏è</div>
                    <div class="impact-value">${impact.key_metrics.productivity_hours_lost}</div>
                    <div class="impact-label">Hours Lost Per Year</div>
                </div>
                <div class="impact-card">
                    <div class="impact-icon">üí°</div>
                    <div class="impact-value">${impact.key_metrics.potential_savings}</div>
                    <div class="impact-label">Potential Savings</div>
                </div>
                <div class="impact-card">
                    <div class="impact-icon">üìà</div>
                    <div class="impact-value">${impact.key_metrics.roi}</div>
                    <div class="impact-label">Expected ROI</div>
                </div>
            `;
            document.getElementById('impactSummary').innerHTML = impactHTML;

            // Display charts
            displayCostChart(impact.visualization_data.cost_breakdown);
            displayROIChart(impact.visualization_data.roi_projection);

            // Display recommendations
            displayRecommendations(data);

            // Show results
            document.getElementById('results').style.display = 'block';

            // Scroll to results
            document.getElementById('results').scrollIntoView({ behavior: 'smooth' });
        }

        function displayCostChart(costData) {
            const ctx = document.getElementById('costChart').getContext('2d');

            if (costChart) {
                costChart.destroy();
            }

            costChart = new Chart(ctx, {
                type: 'doughnut',
                data: {
                    labels: costData.labels,
                    datasets: [{
                        data: costData.values,
                        backgroundColor: [
                            'rgba(99, 102, 241, 0.8)',
                            'rgba(139, 92, 246, 0.8)',
                            'rgba(196, 132, 252, 0.8)'
                        ],
                        borderWidth: 0
                    }]
                },
                options: {
                    responsive: true,
                    maintainAspectRatio: false,
                    plugins: {
                        title: {
                            display: true,
                            text: 'Cost Breakdown by Category',
                            font: { size: 16 }
                        },
                        legend: {
                            position: 'bottom',
                            labels: {
                                padding: 20,
                                font: { size: 14 }
                            }
                        }
                    }
                }
            });
        }

        function displayROIChart(roiData) {
            const ctx = document.getElementById('roiChart').getContext('2d');

            if (roiChart) {
                roiChart.destroy();
            }

            roiChart = new Chart(ctx, {
                type: 'line',
                data: {
                    labels: roiData.months.map(m => `Month ${m}`),
                    datasets: [{
                        label: 'Cumulative Savings',
                        data: roiData.savings,
                        borderColor: '#22c55e',
                        backgroundColor: 'rgba(34, 197, 94, 0.1)',
                        fill: true,
                        tension: 0.4
                    }]
                },
                options: {
                    responsive: true,
                    maintainAspectRatio: false,
                    plugins: {
                        title: {
                            display: true,
                            text: 'Projected Savings Over 12 Months',
                            font: { size: 16 }
                        }
                    },
                    scales: {
                        y: {
                            beginAtZero: true,
                            ticks: {
                                callback: function(value) {
                                    return '$' + value.toLocaleString();
                                }
                            }
                        }
                    }
                }
            });
        }

        function displayRecommendations(data) {
            const biomarkers = data.health_analysis.extracted_biomarkers || {};
            const recs = data.combined_recommendations || [];

            let recHTML = '';

            // Add specific biomarker insights
            if (biomarkers.vitamin_d) {
                recHTML += `
                    <div class="recommendation-item">
                        <strong>üåû Vitamin D Status:</strong> ${biomarkers.vitamin_d.value} ${biomarkers.vitamin_d.unit}
                        ${data.health_analysis.analysis?.vitamin_d?.recommendation || ''}
                    </div>
                `;
            }

            // Add general recommendations
            recs.forEach(rec => {
                recHTML += `<div class="recommendation-item">${rec}</div>`;
            });

            document.getElementById('recommendations').innerHTML = `
                <h3 style="margin-bottom: 1.5rem; color: #e2e8f0;">Your Personalized Action Plan</h3>
                ${recHTML}
            `;
        }

        function loadSampleData() {
            document.getElementById('labReport').value = `Lab Results - Annual Health Check
Date: ${new Date().toLocaleDateString()}

Complete Blood Count (CBC):
WBC: 7.2 x10¬≥/¬µL
Neutrophils: 68%
Lymphocytes: 22%
Monocytes: 7%
Eosinophils: 2%
Basophils: 1%

Chemistry Panel:
Glucose: 95 mg/dL
Vitamin D (25-OH): 18 ng/mL [Low]

Notes: Patient reports fatigue and difficulty concentrating`;
        }
    </script>
</body>
</html>
```
